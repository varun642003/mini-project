# finding ideal learning rate

```python
model=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(200,200)),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(11,activation='softmax')
])

model.compile(loss=tf.keras.losses.CategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(lr=0.2),
              metrics=['accuracy'])

lr_scheduler= tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

lr_find_history=model.fit(X1_train_norm,tf.one_hot(y_train,depth=11),epochs=25,
                          validation_data=(X1_test_norm,tf.one_hot(y_test,depth=11)),
                          callbacks=[lr_scheduler])
```

```python
lrs = 1e-3 * (10**(tf.range(25)/20))
plt.semilogx(lrs,lr_find_history.history['loss'])
plt.xlabel('learning rate')
plt.ylabel('loss')
```
